```python
import pandas as pd
data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)
import matplotlib.pyplot as plt

#1.Line chart(Trend)

plt.plot(df['Days'], df['Study_Hours'], marker='o')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('dhruv')
plt.show()
```

```python
#Advanced line chart
plt.figure(figsize=(7,4))#width 7 inch and height 4 inch
plt.plot(df['Days'], df['Study_Hours'], marker='o',color='green', label='Study Hours')

plt.plot(df['Days'], df['Marks'], marker='s',color='red', label='Marks')

plt.title("dhruv")
plt.xlabel("Days")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
import pandas as pd
data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)
import matplotlib.pyplot as plt

#2. Bar Chart (comparison)

plt.bar(df['Days'], df['Marks'],color='red',edgecolor='black')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.title('dhruv')
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.show()
```

```python
#Advanced bar graph
bars = plt.bar(df['Days'], df['Marks'],color='red',edgecolor='black')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, str(y), ha='center', va='bottom')

plt.xlabel('Days')
plt.ylabel('Marks')
plt.title('dhruv')
plt.grid(axis='y')#add horizontal grid for readability
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(len(df['Days']))  #positions for bars
width=0.36                    # width of each bar
plt.bar(x-width/2,df['Study_Hours'],width,label='Study_Hours',color='green')
plt.bar(x+width/2,df['Marks'],width,label='Marks',color='pink')
plt.xticks(x,df['Days']) #mon tues jo neeche likhe aa rhe hai voh uske vajah saiaa rhe hai varna 1,2,3 likhe aate
plt.xlabel('Days')
plt.ylabel('Hours')
plt.title('dhruv')
plt.legend()
plt.show()
```

```python
#histogram

import matplotlib.pyplot as plt
import numpy as np
#Example data (more values for better histogram)
marks=[50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]
#Calclate mean
mean_value=np.mean(marks)
plt.figure(figsize=(7,4))
plt.hist(marks,bins=3,color='skyblue',edgecolor='black')
#Mean line
plt.axvline(mean_value,color='red',linestyle='--',linewidth=2,label='Mean')
plt.title("dhruv")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)

#conditional scatter plot
#example category
df['Result']=['Pass','Fail','Pass','Pass','Fail']
plot_colors = ['red' if r=='Fail' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'],c=plot_colors)
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('dhruv')
plt.show()
```
